# Cálculo - Tarefa 02

1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  
2- Calcule essa área de forma analítica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
f = lambda x: x**3

# Limites do intervalo
a = 1
b = 2

# Número de retângulos
n = 20

# Calculando a largura do retângulo
delta_x = (b - a) / n

# Calculando as alturas dos retângulos
x_points = np.linspace(a + delta_x/2, b - delta_x/2, n)
y_values = f(x_points)

# Calculando a área dos retângulos e somando-as
approx_integral = np.sum(y_values * delta_x)

# Valores para plotar a função
x = np.linspace(a, b, 100)
y = f(x)

# Plot da função
plt.plot(x, y, label='f(x) = x^3', color='blue')

# Plot dos retângulos
plt.bar(x_points, y_values, width=delta_x, align='center', alpha=0.5, label='Retângulos', color='red')

# Configuração do gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximação Numérica da Integral de x^3')
plt.legend()

# Mostrar o valor da aproximação numérica
plt.annotate(f'Aproximação = {approx_integral:.4f}', xy=(1.2, 1.5), fontsize=12)

# Mostrar o gráfico
plt.show()
```

### **Para calcular a integral de forma analítica, precisamos usar uma fórmula matemática para integrar a função x^3. No caso, podemos usar a seguinte fórmula:**

<br>
<br>

$$ \int x^n \, dx = \frac{x^{n + 1}}{n + 1} + C $$

<br>
<br>
<br>

### **Aplicando essa fórmula, obtemos o seguinte resultado:**
<br>
$$ \int_1^2 x^3 \, dx = \left[ \frac{x^4}{4} \right]_1^2 = \frac{16}{4} - \frac{1}{4} = \boxed{3.375} $$
